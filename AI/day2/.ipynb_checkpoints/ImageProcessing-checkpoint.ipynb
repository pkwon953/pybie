```python
train = cv.imread("./day1/zebra.jpg")
print(train.shape)
b,g,r = cv.split(train)
test =train.copy()
hb = plt.hist(b.ravel(),256,[0,256])#tuple객체로 반환
hg= plt.hist(g.ravel(),256,[0,256])
hr = plt.hist(r.ravel(),256,[0,256])
max_b =(np.where(hb[0]==np.max(hb[0])))[0][0]
print(max_b)
max_g =(np.where(hg[0]==np.max(hg[0])))[0][0]
max_r =(np.where(hr[0]==np.max(hr[0])))[0][0] ## 각 채널의 평균값 구해보기
```

```python
for r in range(train.shape[0]):
    for c in range(train.shape[1]):
        if np.array_equal(test[r,c,:],[max_b,max_g,max_r]):
            test[r,c,:] = 0


plt.imshow(cv.cvtColor(test,cv.COLOR_BGR2RGB))
```

```python
lut1 = [x**x for x  in range(10)] #null transform
print(lut1)
lut2 = [x+5 for x in range(10)]
print(lut2)
lut3 = [x+5 if x < 4 else 9 for x in range(10)]
print(lut3)
def addlut(op) : 
    return[x+op if x+op<255  else 255 for x in range(256)]
def mullut(op) : 
    return[x*op if x*op<255  else 255 for x in range(256)]

print(addlut(5))
lut4 = [np.min((x+5,9)) for x in range(10)]
print(lut4)
```

```python
def mullut(op) : 
    return np.array([x*op if x*op<255 else 255 
                     for x in range(256)]).astype('uint8') #이미지 크기와는 상관없이 돌아가는 명령
#모든 타입이 넘파이여야 계산 가능하다.
src2 = cv.LUT(img,mullut(1.2))
# print(src2)
plt.imshow(src2,cmap='gray')
plt.show()
src3 = cv.LUT(img,mullut(2.0))
plt.imshow(src3,cmap='gray')
plt.show()
all = np.hstack([cv.LUT(img, mullut(0.7)),
           cv.LUT(img, mullut(1.2)),
           cv.LUT(img, mullut(2.0))])
plt.imshow(all,cmap='gray')
```

```python
def binlut(op) : 
    return np.array([ 255 if x>=op else 0 
                     for x in range(256)]).astype('uint8')
def binlut(op) : 
    return np.array([ (x>op)*255 
                     for x in range(256)]).astype('uint8')
def invlut(op) : 
    return np.array([ 255-x  
                     for x in range(256)]).astype('uint8')
def poslut(l) : 
    return np.array([int(x/l)*l
                     for x in range(256)]).astype('uint8')
def gammalut(gamma) : 
    invGamma = 1.0 / gamma
    return np.array([((i/255.0)**invGamma)*255
                     for i in np.arange(0,256)]).astype('uint8')

# src4 = cv.LUT(img,binlut(120))
# src5 = cv.LUT(img,invlut(120))
# src6 = cv.LUT(img,poslut(50))
# src7 = cv.LUT(img,gammalut(2.2))
# all1 = np.hstack([src4,src5,src6,src7])
# plt.imshow(all1,cmap='gray')
```

```python
img = cv.imread('gamma.jpg')
col = cv.LUT(img,gammalut(0.5))
cv.putText(col, "gamma=0.5", (10, 30),
            cv.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 3)
cv.imwrite('g1.jpg',col)

col = cv.LUT(img,gammalut(1.5))
cv.putText(col, "gamma=1.5", (10, 30),
            cv.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 3)
cv.imwrite('g2.jpg',col)

col = cv.LUT(img,gammalut(2.5))
cv.putText(col, "gamma=2.5", (10, 30),
            cv.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 3)
cv.imwrite('g3.jpg',col)
```

```python
gm = [0.5, 1.2, 2.5, 3.0]
for i in range(len(gm)):
    col = cv.LUT(img,gammalut(gm[i]))
    cv.putText(col, f"gamma={gm[i]}", (10, 30),
               cv.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 3)
    cv.imwrite(f'g{i+1}.jpg',np.hstack([img,col]))
```
